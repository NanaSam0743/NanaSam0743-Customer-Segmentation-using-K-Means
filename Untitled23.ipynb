```python
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Add cluster column to original DataFrame
df['Cluster'] = clusters
```

```python
# Plot cluster centers
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1],
            c='black', marker='X', s=200, label='Centroids')
```

```python
# Step 1: Your cluster label dictionary
cluster_labels = {
    0: 'Average People',
    1: 'High In , High Sp',
    2: 'Low In , High Sp',
    3: 'High In , Low Sp',
    4: 'Low In , Low Sp'
}

# Step 2: Map the labels to the Cluster column
df['Segment'] = df['Cluster'].map(cluster_labels)

# (Optional) Step 3: Drop the numeric cluster column if you want only Segment
# df.drop(columns='Cluster', inplace=True)

# Step 4: Display the final DataFrame
df.head()
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))
# Use the correct column names from the DataFrame
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Segment', palette='Set2', s=100)
plt.title('Customer Segmentation')
plt.xlabel('Annual Income (k$)') # Update x-axis label
plt.ylabel('Spending Score (1-100)') # Update y-axis label
plt.show()
```

```python
import pickle

# Save model
with open('kmeans_model.pkl', 'wb') as file:
    pickle.dump(kmeans, file)

# Save DataFrame with clusters
df.to_csv('clustered_customers.csv', index=False)
```
